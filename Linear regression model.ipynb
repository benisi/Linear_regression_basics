```
# used for data analysis and manipulation
import pandas
# It is used for scientific computation like working with arrays and algebra
import numpy
# used for creating visualization
import matplotlib.pyplot as pyplot
#it is used to solves machine learning problems like regression, clustering etc...
from sklearn import linear_model
```

```
dataframe = pandas.read_csv("home_price_sample.csv")
dataframe
```

```
%matplotlib inline
pyplot.xlabel('Area (sqr ft)')
pyplot.ylabel('Price (US$)')
pyplot.scatter(dataframe.area, dataframe.price, color='red', marker="*")
```

```
reg = linear_model.LinearRegression()
reg.fit(dataframe[['area']], dataframe.price)
reg.predict([[3300]])
```
